```python
i=("../Data/stormdata.zip")

zf = ZipFile(i)
    
name = zf.namelist()[0]

print(name)

df_storm = pd.read_csv(zf.open(name))
```

```python
df_storm=df_storm.rename(columns={ "Category_st": "Category" })
```

```python
df_storm.head()
```

```python
df_storm_edited=df_storm[['Station',
 'Sector',
 'Location',  
 "Category",
 '#YY',
 'MM',
 'Date','WVHT(m)', 'R WSP(T m/s)', 'WTMP(Cel)']]
```

```python
df_storm_edited.head()
```

```python
df_edited.head()
```

```python
final=df_edited.append(df_storm_edited)
```

```python
final=final.dropna()
```

```python
target = final["Category"].astype(str)
```

```python
data = final[["R WSP(T m/s)", "WTMP(Cel)", "WVHT(m)"]]
feature_names = data.columns
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, train_size=0.80, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
